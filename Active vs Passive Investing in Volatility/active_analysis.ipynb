```python
active_tickers = [
   "RWMGX",
    "AWSHX",
   "RICGX",
    "ANCFX",
    "AIVSX",
   "FCTDX",
   "RFNGX",
   "FCLKX",
   "FSAKX",
    "VPMCX",
    "LCCAX",
    "FRDPX",
   "GQLIX",
    "HAIAX",
    "MSIGX",
   "JUEAX",
   "JDEAX",
    "PRBLX",
   "RTLAX",
   "CMEUX",
   "CUSUX",
    "PRDGX",
    "PRCOX",
    "VHCOX",
    "VDIGX",
    "VQNPX",
   "BTCLX"
]
```

```python
today = datetime.date.today().strftime('%Y-%m-%d')
active = yf.Tickers(active_tickers).history(start="1990-01-01", end=today)
```

```python
active = active["Close"]
```

```python
earliest_date = active.notnull().all(axis=1).idxmax()
earliest_date
```

```python
active = active.reset_index()
```

```python
active.columns
```

```python
active[(active["Date"] > "2000-01-01") & (active["Date"] < "2000-01-14")].isna().sum()
```

```python
active = active[active["Date"] >= "1998-10-30"]
active = active.rename(columns={"Date": "date"})
```

```python
active['year'] = active['date'].dt.year
active_annual = active.groupby('year').last().drop(columns=['date']).reset_index()
```

```python
active_annual_melted = pd.melt(active_annual, id_vars=['year'], value_vars=active_tickers)
```

```python
active_annual_melted.head()
```

```python
_ = sns.lineplot(data=active_annual_melted, x="year", y="value", hue="Ticker")
```

```python
active_annual_returns = active_annual.pct_change()
active_annual_returns["year"] = active_annual["year"].iloc[1:].astype(int)
```

```python
active_annual_returns["year"].isna().sum()
```

```python
active_annual_returns_melted = pd.melt(active_annual_returns, id_vars=['year'], value_vars=active_tickers)
```

```python
active_annual_returns.columns
```

```python
ax = sns.lineplot(data=active_annual_returns_melted,
                 x=active_annual_returns_melted["year"].iloc[1:], 
                 y="value",
                 hue="Ticker")
sns.move_legend(ax, "upper left", bbox_to_anchor=(1, 1))
```

```python
active_annual_returns["mean"] = active_annual_returns[active_tickers].mean(axis=1)
```

```python
active_annual_returns.head()
```

```python
ax = sns.lineplot(data=active_annual_returns,
                 x="year", 
                 y="mean")
```

```python
active['month_year'] = active['date'].dt.to_period('M')
active_monthly = active.groupby('month_year').last().drop(columns=['date']).reset_index()
```

```python
active_monthly_melted = pd.melt(active_monthly, id_vars=['month_year'], value_vars=active_tickers)
```

```python
_ = sns.lineplot( 
                 x=active_monthly_melted["month_year"].dt.to_timestamp(), 
                 y=active_monthly_melted["value"], 
                 hue=active_monthly_melted["Ticker"])
```

```python
active_monthly_returns = active_monthly[active_tickers].pct_change()
active_monthly_returns["month_year"] = active_monthly["month_year"].iloc[1:].astype(int)
```

```python
active_monthly_returns_melted = pd.melt(active_monthly_returns, id_vars=['month_year'], value_vars=active_tickers)
```

```python
ax = sns.lineplot(data=active_monthly_returns_melted,
                 x=active_monthly_returns_melted["month_year"].iloc[1:], 
                 y="value",
                 hue="Ticker")
sns.move_legend(ax, "upper left", bbox_to_anchor=(1, 1))
```
